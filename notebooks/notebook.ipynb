### Section 1: Visualize `MaierStein_pts.csv`
which can be found in the external data directory.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from MaierSteinSolver.config import EXTERNAL_DATA_DIR
```

```python
pts = np.loadtxt(f'{EXTERNAL_DATA_DIR}/MaierStein_pts.csv', delimiter=',', dtype=float)
```

```python
fig, ax = plt.subplots()
ax.scatter(pts[:,0], pts[:,1], s=1)
ax.set_xlabel("x-axis")
ax.set_ylabel("y-axis")
ax.set_title("FEM inputs: 4224 points")
fig.show()
```
